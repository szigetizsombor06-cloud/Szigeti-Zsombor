```python
# Install into the active Jupyter kernel environment
import sys
print("Kernel Python:", sys.executable)

try:
	import yfinance as yf
	print("yfinance already available")
except ImportError:
	# Use Jupyter-aware magic so it installs into this kernel
	import sys
	if 'ipykernel' in sys.modules:
		# %pip is a magic handled by IPython to target the current kernel
		import IPython
		ip = IPython.get_ipython()
		ip.run_line_magic('pip', 'install yfinance matplotlib --quiet')
		import yfinance as yf
		print("yfinance installed and imported")
	else:
		# Fallback if not running in IPython kernel
		import subprocess
		subprocess.check_call([sys.executable, '-m', 'pip', 'install', 'yfinance', 'matplotlib'])
		import yfinance as yf

# quick test
import matplotlib.pyplot as plt

data = yf.download("AAPL", period="1mo")
print("Downloaded rows:", len(data))
data["Close"].plot(title="AAPL Close - 1mo")
plt.show()
```
